```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2


chemin_photo = "test/trail_only/test__.jpg"
chemin_photo =  "test/trail_only/prog/prise1.jpg"
img = mpimg.imread(chemin_photo)
imgc = img.copy()
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.show()

B, G, R = cv2.split(img)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax = axes.ravel()
ax[0].imshow(B, cmap=plt.cm.gray)
ax[1].imshow(G, cmap=plt.cm.gray)
ax[2].imshow(R, cmap=plt.cm.gray)
plt.show()

img = cv2.cvtColor(imgc, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(img)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax = axes.ravel()
ax[0].imshow(H, cmap=plt.cm.gray)
ax[1].imshow(S, cmap=plt.cm.gray)
ax[2].imshow(V, cmap=plt.cm.gray)
plt.show()


seuil_otsu = cv2.threshold(S,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1]
plt.figure(figsize=(8,8))
plt.imshow(seuil_otsu)
plt.show()


seuil_adapt = cv2.adaptiveThreshold(S, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C ,cv2.THRESH_BINARY, 5, 0 )
plt.figure(figsize=(8,8))
plt.imshow(seuil_adapt, cmap=plt.cm.gray)
plt.show()

ret,thresh = cv2.threshold(S,130,255,cv2.THRESH_BINARY)
plt.figure(figsize=(8,8))
plt.imshow(thresh, cmap=plt.cm.gray)
plt.show()

img = cv2.cvtColor(imgc, cv2.COLOR_BGR2HSV_FULL)
H, S, V = cv2.split(img)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax = axes.ravel()
ax[0].imshow(H, cmap=plt.cm.gray)
ax[1].imshow(S, cmap=plt.cm.gray)
ax[2].imshow(V, cmap=plt.cm.gray)
plt.show()

img = cv2.cvtColor(imgc, cv2.COLOR_BGR2HLS)
H, S, V = cv2.split(img)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax = axes.ravel()
ax[0].imshow(H, cmap=plt.cm.gray)
ax[1].imshow(S, cmap=plt.cm.gray)
ax[2].imshow(V, cmap=plt.cm.gray)
plt.show()

img = cv2.cvtColor(imgc, cv2.COLOR_BGR2BGR555)
H, S = cv2.split(img)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
ax = axes.ravel()
ax[0].imshow(H, cmap=plt.cm.gray)
ax[1].imshow(S, cmap=plt.cm.gray)
plt.show()

img = cv2.cvtColor(imgc, cv2.COLOR_BGR2BGR565)
H, S= cv2.split(img)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
ax = axes.ravel()
ax[0].imshow(H, cmap=plt.cm.gray)
ax[1].imshow(S, cmap=plt.cm.gray)
plt.show()

img = cv2.cvtColor(imgc, cv2.COLOR_BGR2BGRA)
H, S, V, A = cv2.split(img)
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
ax = axes.ravel()
ax[0].imshow(H, cmap=plt.cm.gray)
ax[1].imshow(S, cmap=plt.cm.gray)
ax[2].imshow(V, cmap=plt.cm.gray)
ax[3].imshow(A, cmap=plt.cm.gray)
plt.show()
```
