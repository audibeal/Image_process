```python
import numpy as np

from skimage.transform import hough_line, hough_line_peaks
from skimage.feature import canny
from skimage import data

import matplotlib.pyplot as plt
from matplotlib import cm


# Constructing test image
image  = mpimg.imread('test/original/111.jpg')
imgc = image.copy()
img_ = np.zeros_like(imgc)
#image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#image = cv2.Canny(image,50,150,apertureSize = 3)
image = cv2.Canny(image,100,200)
image = cv2.dilate(image, None, iterations=3)


# Classic straight-line Hough transform
# Set a precision of 0.5 degree.
tested_angles = np.linspace(-np.pi / 2, np.pi / 2, 360)
h, theta, d = hough_line(image, theta=tested_angles)
print(len(theta))
print(len(d))
# Generating figure 1
fig, axes = plt.subplots(1, 3, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(image, cmap=cm.gray)
ax[0].set_title('Input image')
ax[0].set_axis_off()

ax[1].imshow(np.log(1 + h),
             extent=[np.rad2deg(theta[-1]), np.rad2deg(theta[0]), d[-1], d[0]],
             cmap=cm.gray, aspect=1/1.5)
ax[1].set_title('Hough transform')
ax[1].set_xlabel('Angles (degrees)')
ax[1].set_ylabel('Distance (pixels)')
ax[1].axis('image')

ax[2].imshow(image, cmap=cm.gray)
origin = np.array((0, image.shape[1]))
for _, angle, dist in zip(*hough_line_peaks(h, theta, d, threshold=1600)):
    y0, y1 = (dist - origin * np.cos(angle)) / np.sin(angle)
    ax[2].plot(origin, (y0, y1), '-r')
l = []
for _, theta, rho in zip(*hough_line_peaks(h, theta, d, threshold=1600)):
    l.append((theta, rho))
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 4496*(-b))
    y1 = int(y0 + 4496*(a))
    x2 = int(x0 - 4496*(-b))
    y2 = int(y0 - 4496*(a))

    cv2.line(img_,(x1,y1),(x2,y2),(255,255,255),2)
    

  
ax[2].set_xlim(origin)
ax[2].set_ylim((image.shape[0], 0))
ax[2].set_axis_off()
ax[2].set_title('Detected lines')

plt.tight_layout()
plt.show()


ListeClassesEquivalence = []

eps = 0.2
for theta, rho in l:
    # je regarde si c'est dans une classe d'equive
    ce = []
    for i, classe in enumerate(ListeClassesEquivalence):
        for t, r in classe:
            if abs(t/theta - 1) < eps:
                ce.append(i)
                break
    if len(ce) == 0:#pas dans un classe d'equiv
        ListeClassesEquivalence.append([(theta, rho)])
    else:
        ce.reverse()
        new_classe = [(theta, rho)]
        for i in ce:
            classe = ListeClassesEquivalence.pop(i)
            for rho, theta in classe:
                new_classe.append(( theta, rho))
        ListeClassesEquivalence.append(new_classe)
            
print(len(ListeClassesEquivalence))
print(ListeClassesEquivalence)

# je vais garder les 2 classes d'equiv les plus grandes
ListeClassesEquivalence = sorted(ListeClassesEquivalence, key=lambda classe: -len(classe))

print(ListeClassesEquivalence)
ListeClassesEquivalence = ListeClassesEquivalence[:2]
print(ListeClassesEquivalence)

def function__(liste):
    r1 = (0,0)
    r2 = (0,0)
    g = 0
    for i in range(len(liste)):
        for j in range(i+1, len(liste)):
            gg = abs(liste[i][1] - liste[j][1])
            if gg > g:
                g = gg
                r1 = liste[i]
                r2 = liste[j]
    return r1, r2

d1, d2 = function__(ListeClassesEquivalence[0])
d3, d4 = function__(ListeClassesEquivalence[1])
kk = [d1, d2, d3, d4]

img___ = img.copy()
for theta, rho in kk:
    print(theta, rho)
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 4496*(-b))
    y1 = int(y0 + 4496*(a))
    x2 = int(x0 - 4496*(-b))
    y2 = int(y0 - 4496*(a))

    cv2.line(img___,(x1,y1),(x2,y2),(255,0,0),2)
    
plt.figure(figsize=(8,8))
plt.imshow(img___)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

import matplotlib.image as mpimg
methode = cv2.ADAPTIVE_THRESH_GAUSSIAN_C
#methode = cv2.ADAPTIVE_THRESH_MEAN_C
v1 = 3
img = mpimg.imread('test/original/111.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (5,5), 0)
#edges = cv2.Canny(img,100,200)
thresh = cv2.adaptiveThreshold(gray, 255, methode ,cv2.THRESH_BINARY_INV, v1, 2 )
#thresh = cv2.dilate(thresh, None, iterations=1)
thresh = cv2.Canny(img,100,200)
thresh = cv2.dilate(thresh,  np.ones((5,5),np.uint8), iterations=10)
plt.imshow(thresh,cmap=plt.cm.gray )
```

```python
imgc = img.copy()
contours, hierarchy=cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_grille=None
maxArea=0
for c in contours:
    #cv2.drawContours(imgc, [c], 0, (0, 255, 0), 2)
    area=cv2.contourArea(c)
    if area>100:
        peri=cv2.arcLength(c, True)
        polygone=cv2.approxPolyDP(c, 0.1*peri, True)
        if area>maxArea and len(polygone)==4:
            contour_grille=polygone
            maxArea=area
if contour_grille is not None:
    cv2.drawContours(imgc, [contour_grille], 0, (0, 255, 0), 2)
    
plt.imshow(imgc)
```
